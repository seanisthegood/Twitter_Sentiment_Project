# Weekly Tips Report Generator

## Business Problem

At restaruant Esme, employee tips are entered into a Google Sheet to give an output for processing to give the pooled tipped amounts to a payment processor. The sheet is designed for the input, but numbers must be manually entered into the payment processor. This is short, but tedious work.

Additionaly, employees want a weekly sumation of their tips in order to see their tips for each shift. The tips each worker receives in a "pooled house" are not the tips they take home. A calculatino is made based on their "points" - a set percentage assigned based on role. A may take in 300 in tips, but share that with a back server and a bartender. That server may take home 250 in tips.  This is a more or less easy pen and paper calculation, but why not use what we already have with the some python to generate a weekly report that -

-Gives each worker the tips they made for each shift
-The tips that their coworkers received
-Does not let the server see tips from days they did not work

## Functions to Create Smallers Dataframes from Larger Dataframes

```python
def to_dictionary(frame):
    #Function that transforms large dataframe to dictionary.
    frame.columns = frame.iloc[0]
    
    frame.fillna(0,inplace=True)
    
    frame.drop(2)
    
    frame.drop(2,inplace=True)
    frame.reset_index(inplace=True)
    frame.drop('index',axis=1,inplace=True)
    frame.rename(columns={0:'Employee'},inplace=True)
    frame[' Payroll ']=frame[' Payroll '].str.replace('$',"")
    frame[' Payroll ']=frame[' Payroll '].str.replace('-',"0")
    frame[' Payroll ']=frame[' Payroll '].str.replace(',',"")
    
    frame.fillna(0,inplace=True)
    frame = frame.astype({'Points':'float','Hours':'float','Cut':'float',' Payroll ':'float'})
    
    frame_sorted = frame[frame['Hours']>0]
    frame_sorted.reset_index(inplace=True,drop=True)
    
    frame_sorted = frame_sorted.to_dict(orient='')
    return frame_sorted
```

```python
def to_frame(frame):
    #function that transforms large dataframe into smaller dataframe per day.
    frame.columns = frame.iloc[0]
    
    frame.fillna(0,inplace=True)
    
    frame.drop(2)
    
    frame.drop(2,inplace=True)
    frame.reset_index(inplace=True)
    frame.drop('index',axis=1,inplace=True)
    frame.rename(columns={0:'Employee'},inplace=True)
    frame.replace('')
    frame.replace('#REF!',0,inplace=True)
    #Removing pesky characters from the Payroll Column
    frame[' Payroll ']=frame[' Payroll '].str.replace('$',"")
    frame[' Payroll ']=frame[' Payroll '].str.replace('-',"0")
    frame[' Payroll ']=frame[' Payroll '].str.replace(',',"")
    #Cleaning out extra characters from the Employee dataframe
    frame['Employee'] = frame['Employee'].str.findall(r'[^\s,]+').str.join(', ')
    
    frame.fillna(0,inplace=True)
    frame = frame.astype({'Points':'float','Hours':'float','Cut':'float',' Payroll ':'float'})
    
    frame_sorted = frame[frame['Hours']>0]
    frame_sorted.reset_index(inplace=True,drop=True)
    
    return frame_sorted
```

### Report Functions

The day report function will extract the tips for the given employee and include the other employees tips for the days along with the totals for the day.

```python
#create a function that extracts the index for the days the employee worked
def day_report(name,df):
    #extract the index for the give employee
    idx = df[df['Employee']==name].index
    #list comp to get the first part of the index tupples
    idx_list = [x[0] for x in idx]
    #return first part of the index from the tupples
    return df.loc[idx_list,:]
```

The ind report function will extract only the given employees tips for the day along with the total tips for the day. 

```python
#function that returns individual's tips and the total tips for the day
def ind_report(name,df):
    #extract the index for the give employee
    idx = df[df['Employee']==name].index
    #list comp to get the first part of the index tupples
    idx_list = [x[0] for x in idx]
    #return first part of the index from the tupples
    df =df.loc[idx_list,:]
    return df[(df['Employee']==name) | (df['Employee']=='TOTALS')]
#               test[(test['Employee']=='Hart, Sean') | (test['Employee']=='TOTALS')]
```

### Plotting Functions

```python
def show_values(axs, orient="v", space=.01):
    #This is a function that adds the numeric value label to the top of the bar graph
    def _single(ax):
        if orient == "v":
            for p in ax.patches:
                _x = p.get_x() + p.get_width() / 2
                _y = p.get_y() + p.get_height() + (p.get_height()*0.01)
                value = '${:.2f}'.format(p.get_height())
                ax.text(_x, _y, value, ha="center") 
        elif orient == "h":
            for p in ax.patches:
                _x = p.get_x() + p.get_width() + float(space)
                _y = p.get_y() + p.get_height() - (p.get_height()*1.2)
                value = '${:.2f}'.format(p.get_width())
                ax.text(_x, _y, value, ha="left")

    if isinstance(axs, np.ndarray):
        for idx, ax in np.ndenumerate(axs):
            _single(ax)
    else:
        _single(axs)
```

```python
def plot_file(emp):
    #Loop that goes through the employee list
    sns.set_style("whitegrid")
    plt.figure(figsize = (10,6))
    plt.style.use('fivethirtyeight')
    #creating an iteration for each employee and extracting the tip information from the dictionary
    ax = sns.barplot(data=emp_dict[emp]['tips'].reset_index(),x='shift',y=' Payroll ',hue='Employee',estimator=sum,ci=None)
    ax.set_ylabel('Tips $')
    ax.set_xlabel('Shift')
    plt.legend(loc='upper left')
    show_values(ax)
    #Saving each image to be distributed in an email
    emp_dict[emp]['file_name']=f'plot{emp}.png'
    return plt.savefig(f'plot{emp}.png')
```
